Write a program to count all the word types that occur at least 75 times in the Brown Corpus. How many of them are there?

```python
import nltk
from nltk.corpus import brown
from nltk import FreqDist
```

```python
nltk.download('brown')

words = brown.words()

freq_dist = FreqDist(words)

word_types_at_least_75 = [word for word, count in freq_dist.items() if count >= 75]

for word_type in word_types_at_least_75:
    print(f"{word_type}: {freq_dist[word_type]}")

print("\nTotal number of word types occurring at least 75 times:", len(word_types_at_least_75))
```

Read in the texts of the State of the Union addresses, using the state_union corpus reader. Count occurrences of men, women, and people in each document. What has happened to the usage of these words over time?

```python
import nltk
from nltk.corpus import state_union
from nltk import FreqDist
import matplotlib.pyplot as plt

nltk.download('state_union')

files = state_union.fileids()

men_freq = {}
women_freq = {}
people_freq = {}

for file_id in files:
    words = state_union.words(file_id)
    freq_dist = FreqDist(words)

    men_freq[file_id] = freq_dist['men']
    women_freq[file_id] = freq_dist['women']
    people_freq[file_id] = freq_dist['people']

plt.figure(figsize=(12, 6))

plt.plot(men_freq.keys(), men_freq.values(), label='Men')
plt.plot(women_freq.keys(), women_freq.values(), label='Women')
plt.plot(people_freq.keys(), people_freq.values(), label='People')

plt.title('Usage of Words Over Time in State of the Union Addresses')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Write a program to guess the number of syllables contained in a text, making use of the CMU Pronouncing Dictionary. [This one is hard! For some ideas about how to solve it, look at the discussion of the CMU dictionary in the NLTK book. If you haven't done much programming before you'll probably find it very challenging, but try your best.]

```python
import nltk
from nltk.corpus import cmudict
from nltk.tokenize import word_tokenize

nltk.download('cmudict')
nltk.download('punkt')

pronouncing_dict = cmudict.dict()

def count_syllables(word):
    pronunciations = pronouncing_dict.get(word.lower(), [])

    return max([len([ph for ph in pron if any(char.isdigit() for char in ph)]) for pron in pronunciations], default=0)

def estimate_syllables(text):
    words = word_tokenize(text)

    total_syllables = sum(count_syllables(word) for word in words)

    return total_syllables
text = "This is an example sentence to estimate the number of syllables in a text."
syllable_count = estimate_syllables(text)
print(f"Estimated syllable count: {syllable_count}")
```
